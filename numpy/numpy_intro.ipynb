# Array Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
# max/mean/sum/prod

np.max(a1, axis=1)
```

```python
# mean/median/std/var

np.mean(a1)
```

```python
# dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
# log and exponents

np.log()
```
