# Broadcasting 

```python
# same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a)
print(b)

print(a+b)
```

```python
# diff shape

a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print(a+b)
```

```python
# more example 
import numpy as np 
a = np.arange(12).reshape(4,3)
b= np.arange(3)

print(a)
print(b)

print(a+b)
```
