```python
class MyException(Exception): # this is the exception class 
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

class bank:
    def __init__(self, balance):
        self.balance = balance
        
    def withdraw(self, amount):
        if amount < 0:
            raise MyException('amount cannot be negative')
        if self.balance < amount:
            raise MyException('not money')
        self.balance = self.balance - amount

obj = bank(10000)
try:
    obj.withdraw(15000)
except MyException as e:
    print(f"Exception caught: {e}")
else:
    print(obj.balance)
```

```python
class SecurityError(Exception):

    def __init__(self,message):
        print(message)
    
    def logout(self):
        print('logout')

class google:

    def __init__(self, name, email, password, device):
        self.name = name
        self.email = email
        self.password = password
        self.device = device
    
    def login(self, email, password, device):
        if device != self.device:
            raise SecurityError('error')
        if email == self.email and password == self.password:
            print('welcome')
        else:
            print('login error')



obj = google('naveen','naveen8522374@gmail.com','1234','android')
try:
    obj.login('naveen852374@gmail.com','12340','windows')
except SecurityError as e:
    e.logout()
else:
    print(obj.name)
finally:
    print('database connection closed')
```
